```python
import ast
def process_genres(genres_list):
    genres_list = ast.literal_eval(genres_list)
    unique_genres = set()
    processed_genres = []
    for genre in genres_list:
        if '---' not in genre:
            if genre not in unique_genres:
                unique_genres.add(genre)
                processed_genres.append(genre)
    return processed_genres
```

```python
my_df['genres_acousticb_processed'] = my_df['genres_acousticb'].apply(process_genres)
```

```python
#selected genres
selected_genres = ['electronic' , 'rock' , 'funk / soul'  , 'pop'  , 'folk, world, & country' , 'hip hop'  , 'blues' , 'jazz' , 'latin', 'classical']
def process_genres_prcessed(genres_list):
  # genres_list = ast.literal_eval(genres_list)
  genres=[]
  for genre in genres_list:
    if genre in selected_genres:
      genres.append(genre)
  return genres

my_df['genres_acousticb_processed_2'] = my_df['genres_acousticb_processed'].apply(process_genres_prcessed)

```

```python
def further_process(genres_list):
  genres=[]
  map = {
      'funk / soul': 'funk&soul',
      'folk, world, & country': 'folk&country',
      'hip hop': 'hiphop',
      'blues': 'rhythm&blues'
  }
  for genre in genres_list:
    genre=genre.strip()
    if genre in map:
      genres.append(map[genre])
    else:
      genres.append(genre)
  return genres

my_df['genres_acousticb_processed_2'] = my_df['genres_acousticb_processed_2'].apply(further_process)
```

```python
def process_genius(genre):
  map = {
      'rap': 'hiphop',
      'rb': 'rhythm&blues',
      'country': 'folk&country'
  }
  if genre in map:
    genre = map[genre]
  return genre

my_df['genres_genius'] = my_df['genres_genius'].apply(process_genius)
```

```python
from typing_extensions import NewType
def process_self(self_list):
  self_list = ast.literal_eval(self_list)
  new=[]
  map={
      'blues': 'rhythm&blues',
      'country': 'folk&country',
      'folk': 'folk&country',
      'funk': 'funk&soul',
  }
  for gen in self_list:
    if gen in map:
      new.append(map[gen])
    else:
      new.append(gen)
  return new
my_df['genres_self'] = my_df['genres_self'].apply(process_self)
```

```python
from pickle import NEWOBJ_EX
def process_mbrainz(l):
  l = ast.literal_eval(l)
  new=[]
  map = {
   'electronic': 'electronic',
   'rock': 'rock',
   'funk': 'funk&soul',
   'soul': 'funk&soul',
   'pop': 'pop',
   'folk': 'folk&country',
   'country': 'folk&country',
   'hiphop': "hiphop",
   'hip': 'hiphop',
   'hop': 'hiphop',
   'rap': 'hiphop',
   'blues': 'rhythm&blues',
   'rhythm': 'rhythm&blues',
   'latin': 'latin',
   'jazz': 'jazz',
   'classical' :'classical'
  }
  counts = {
      'electronic': 0,
      'rock': 0,
      'funk&soul': 0,
      'pop': 0,
      'folk&country': 0,
      'hiphop': 0,
      'rhythm&blues': 0,
      'latin': 0,
      'jazz': 0,
      'classical': 0
  }

  for c,gen in l:
    for x in gen.split(' '):
      x= x.strip().lower()
      if x in map:
        counts[map[x]]+=1

  new = [x for x,y in counts.items() if y>=2]
  return new

my_df['genres_mb2'] = my_df['genres_mb'].apply(process_mbrainz)
```

```python
def process_mb_genres(genre_list):
  final_gen = []
  # genre_list = ast.literal_eval(genre_list)
  for genre in genre_list:
    x = genre.strip().lower().split(' ')
    for g in x:
      if g in map:
        if map[g] not in final_gen:
          final_gen.append(map[g])
  return final_gen

my_df['genres_mb3'] = my_df['genres_mb2'].apply(process_mb_genres)
```
